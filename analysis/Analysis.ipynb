```python
# Pivot the DataFrame to make 'parameter' values the new columns
df_pivoted = data.pivot_table(
    index='datetimeLocal', 
    columns='parameter', 
    values='value'
)

# Reset the index to make 'datetimeLocal' a regular column
df_pivoted = df_pivoted.reset_index()
```

```python
def calculate_pm25_aqi(Cp):
    """
    Calculates the PM2.5 AQI based on the US EPA breakpoints.
    Cp: PM2.5 concentration in ug/m3 (micrograms per cubic meter).
    """
    # Define AQI breakpoints: (C_Lo, C_Hi, I_Lo, I_Hi)
    breakpoints = [
        (0.0, 12.0, 0, 50),
        (12.1, 35.4, 51, 100),
        (35.5, 55.4, 101, 150),
        (55.5, 150.4, 151, 200),
        (150.5, 250.4, 201, 300),
        (250.5, 500.4, 301, 500)
    ]

    # Handle concentrations outside the defined range
    if Cp < 0.0:
        return 0
    if Cp > 500.4:
        # Returns maximum index for the given standard
        return 500

    # Find the correct interval and calculate AQI
    for C_Lo, C_Hi, I_Lo, I_Hi in breakpoints:
        if C_Lo <= Cp <= C_Hi:
            # Piecewise linear formula: I_p = [(I_Hi - I_Lo) / (C_Hi - C_Lo)] * (C_p - C_Lo) + I_Lo
            if C_Hi == C_Lo: # Safety check
                 return I_Hi
            
            # Use max(Cp, C_Lo) to handle floating point precision near C_Lo
            C_prime = max(Cp, C_Lo)
            
            Ip = ((I_Hi - I_Lo) / (C_Hi - C_Lo)) * (C_prime - C_Lo) + I_Lo
            
            return round(Ip)

    # Fallback, should return NaN for unhandled edge cases
    return np.nan
```

```python
df_pivoted['aqi_estimate'] = df_pivoted['pm25'].apply(calculate_pm25_aqi)
```

```python
df_pivoted.head(10)
```

```python
# ==============================
# 1️⃣ Time Series Plot (AQI over Time)
# ==============================
plt.figure(figsize=(10,5))
plt.plot(df_pivoted["datetimeLocal"], df_pivoted["aqi_estimate"], marker='o', color='r')
plt.title("AQI Trend Over Time")
plt.xlabel("Datetime")
plt.ylabel("AQI Estimate")
plt.grid(True)
plt.show()
```

```python
# ==============================
# 2️⃣ Compare Pollutants over Time
# ==============================
plt.figure(figsize=(10,5))
plt.plot(df_pivoted["datetimeLocal"], df_pivoted["pm1"], label="PM1")
plt.plot(df_pivoted["datetimeLocal"], df_pivoted["pm25"], label="PM2.5")
plt.title("PM1 vs PM2.5 Over Time")
plt.xlabel("Datetime")
plt.ylabel("Concentration (µg/m³)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ==============================
# 3️⃣ Correlation Heatmap
# ==============================
plt.figure(figsize=(8,6))
sns.heatmap(df_pivoted.drop(columns=["datetimeLocal"]).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# ==============================
# 4️⃣ Relationship: Temperature vs AQI
# ==============================
plt.figure(figsize=(6,5))
sns.scatterplot(data=df_pivoted, x="temperature", y="aqi_estimate", hue="relativehumidity", palette="viridis", s=100)
plt.title("Temperature vs AQI (colored by Humidity)")
plt.show()
```

```python
# ==============================
# 5️⃣ Pairplot for Quick EDA
# ==============================
sns.pairplot(df_pivoted.drop(columns=["datetimeLocal"]), diag_kind="kde", corner=True)
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

```python
# Seaborn style
sns.set(style="whitegrid", context="talk")

# Create 3 subplots for PM1, PM2.5, and Temperature
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# AQI vs PM1
sns.regplot(data=df_pivoted, x='pm1', y='aqi_estimate', ax=ax[0],
            scatter_kws={'s':70, 'alpha':0.8}, line_kws={'color':'red'})
ax[0].set_title('AQI vs PM1')
ax[0].set_xlabel('PM1 (µg/m³)')
ax[0].set_ylabel('AQI')

# AQI vs PM2.5
sns.regplot(data=df_pivoted, x='pm25', y='aqi_estimate', ax=ax[1],
            scatter_kws={'s':70, 'alpha':0.8}, line_kws={'color':'red'})
ax[1].set_title('AQI vs PM2.5')
ax[1].set_xlabel('PM2.5 (µg/m³)')
ax[1].set_ylabel('AQI')

# AQI vs Temperature
sns.regplot(data=df_pivoted, x='um003', y='aqi_estimate', ax=ax[2],
            scatter_kws={'s':70, 'alpha':0.8}, line_kws={'color':'red'})
ax[2].set_title('AQI vs um003')
ax[2].set_xlabel('um003')
ax[2].set_ylabel('AQI')

plt.tight_layout()
plt.show()
```
